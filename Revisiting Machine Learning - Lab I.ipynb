*Replace blank space on RECINHSE, RECP3, RECPGVG, RECSWEEP with 'N'.*

```python
cols_lst = ['RECINHSE', 'RECP3', 'RECPGVG', 'RECSWEEP']

for col in cols_lst:
    categorical[col] = categorical[col].apply(lambda x: x.replace(' ', 'N'))
    
print('Unique values of {} are {}'.format('RECINHSE', categorical['RECINHSE'].unique()))
print('Unique values of {} are {}'.format('RECP3', categorical['RECP3'].unique()))
print('Unique values of {} are {}'.format('RECPGVG', categorical['RECPGVG'].unique()))
print('Unique values of {} are {}'.format('RECSWEEP', categorical['RECSWEEP'].unique()))
```

Handling unique values of NOEXCH.

```python
categorical['NOEXCH'].unique()
```

```python
def change_values(row):
    if row == " ":
        return row.replace(' ', 'N')
    elif row in ['0', '1', 0, 1]:
        return np.nan
    else:
        return row
```

```python
categorical['NOEXCH'] = categorical['NOEXCH'].apply(change_values)
categorical['NOEXCH'].unique()
```

```python
categorical['NOEXCH'].isna().sum()
```

**Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest.**

Note that there are a few null values in the column. We will first replace those null values using the code below:

```python
print(categorical['GENDER'].unique())
print(categorical['GENDER'].value_counts())
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
print(categorical['GENDER'].unique())
print(categorical['GENDER'].value_counts())
```

```python
def change_values_gender(row):
    if row in ['U', 'J','C','A']:
        return 'other'
    else:
        return row
```

```python
categorical['GENDER'] = categorical['GENDER'].apply(change_values_gender)
print(categorical['GENDER'].unique())
print(categorical['GENDER'].value_counts())

```
